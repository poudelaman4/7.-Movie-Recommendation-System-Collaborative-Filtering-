```python
def recommend_items(movie_title, top_n=10):
    """
    Recommends movies similar to a given movie title.
    """
    try:
        # Get the similarity scores for the input movie
        sims = movie_sim_df[movie_title].sort_values(ascending=False)

        # Drop the movie itself from the list
        sims = sims.drop(labels=[movie_title])

        # Return the top N most similar movies
        return sims.iloc[:top_n]
    except KeyError:
        print(f"Movie '{movie_title}' not found. Please try a different title.")
        return None

# Let's test it out!
recommended_movies = recommend_items('Star Wars (1977)')

if recommended_movies is not None:
    print("\nTop 10 Recommendations for 'Star Wars (1977)':")
    print(recommended_movies)
```

```python
# 1. Start with the merged DataFrame (df) from our previous steps
# This DataFrame contains all the ratings, users, and movies.

# 2. Divide the data into a training and testing set
# A simple way is to use a 80/20 split, but this can be tricky for recommender systems
# Let's create a more controlled split, ensuring each user has ratings in both sets
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# 3. Rebuild the user-movie matrix using only the TRAINING data
train_user_movie = train_df.pivot_table(index='user_id', columns='title', values='rating')
train_movie_matrix = train_user_movie.fillna(0).T
train_movie_sim_df = pd.DataFrame(cosine_similarity(train_movie_matrix),
                                  index=train_user_movie.columns,
                                  columns=train_user_movie.columns)

# 4. Create a function to recommend based on the user's highest-rated movies
def recommend_for_user(user_id, train_df, train_movie_sim_df, top_k=10):
    # Get the movies the user rated highly (e.g., 4 or 5 stars)
    user_movies = train_df[(train_df['user_id'] == user_id) & (train_df['rating'] >= 4)]['title'].unique()
    
    if len(user_movies) == 0:
        return []

    # Get the recommendations for all their highly rated movies
    all_recommendations = pd.Series(dtype='float64')
    for movie in user_movies:
        if movie in train_movie_sim_df.index:
            sims = train_movie_sim_df[movie].drop(labels=[movie])
            all_recommendations = pd.concat([all_recommendations, sims])

    # Rank the combined recommendations and get the top_k
    return all_recommendations.sort_values(ascending=False).index.unique()[:top_k].tolist()

# 5. Evaluate the recommender using Precision@K
def evaluate_precision_at_k(train_df, test_df, model_func, k=10):
    user_ids = test_df['user_id'].unique()
    total_precision = 0
    
    # Track how many recommendations were made
    total_users_evaluated = 0

    for user_id in user_ids:
        # Get the ground truth movies from the test set
        true_positives = test_df[(test_df['user_id'] == user_id) & (test_df['rating'] >= 4)]['title'].unique()
        
        # Get recommendations from our model
        recommended_movies = model_func(user_id, train_df, train_movie_sim_df, top_k=k)
        
        # We only evaluate users for whom recommendations can be made
        if len(recommended_movies) == 0:
            continue
            
        # Count how many recommendations are relevant (in the test set)
        hits = 0
        for movie in recommended_movies:
            if movie in true_positives:
                hits += 1
        
        # Calculate precision for this user
        precision = hits / len(recommended_movies)
        total_precision += precision
        total_users_evaluated += 1

    # Calculate the average precision over all evaluated users
    if total_users_evaluated > 0:
        avg_precision = total_precision / total_users_evaluated
        print(f"Average Precision@{k}: {avg_precision:.4f}")
    else:
        print("Could not evaluate any users. Check your data split and filtering.")

# 6. Run the evaluation
evaluate_precision_at_k(train_df, test_df, recommend_for_user, k=10)
```

```python
train_df = pd.merge(train, movies, on='movie_id')
train_matrix = train_df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Fill missing with 0 for similarity computation
user_similarity = cosine_similarity(train_matrix.fillna(0))
user_similarity_df = pd.DataFrame(user_similarity,
                                  index=train_matrix.index,
                                  columns=train_matrix.index)
```

We can predict user u’s rating on item i using:

r^ui = (∑v sim(u,v) * rvi) / (∑v |sim(u,v)|)

```python
def predict_rating(user_id, movie_title, train_matrix, user_similarity_df):
    if movie_title not in train_matrix.columns:
        return np.nan
    
    # Get similarity scores for this user to all other users
    sims = user_similarity_df.loc[user_id]
    
    # Ratings of all other users for this movie
    ratings_for_movie = train_matrix[movie_title]
    
    # Mask users who didn’t rate this movie
    mask = ratings_for_movie.notna()
    
    # Weighted average
    numerator = (sims[mask] * ratings_for_movie[mask]).sum()
    denominator = np.abs(sims[mask]).sum()
    if denominator == 0:
        return np.nan
    return numerator / denominator
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

y_true = []
y_pred = []

for _, row in test.iterrows():
    uid = row['user_id']
    mid = row['movie_id']
    rating = row['rating']
    
    movie_title = movies.loc[movies['movie_id']==mid, 'title'].values[0]
    
    pred = predict_rating(uid, movie_title, train_matrix, user_similarity_df)
    if not np.isnan(pred):
        y_true.append(rating)
        y_pred.append(pred)

rmse = sqrt(mean_squared_error(y_true, y_pred))
print("RMSE on test set:", rmse)
```
